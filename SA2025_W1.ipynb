### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd

# Read the CSV file
cars_df = pd.read_csv('Cars.csv')
def extract_base_model(name):
    # Remove year suffixes like numbers, roman numerals, trim levels
    import re
    base = re.sub(r'\s+\d{4}$', '', name)  # Remove 4-digit years
    base = re.sub(r'\s+[ivx]+$', '', base, flags=re.IGNORECASE)  # Remove roman numerals
    base = re.sub(r'\s+[a-z]$', '', base, flags=re.IGNORECASE)  # Remove single letter suffixes
    base = re.sub(r'\([^)]*\)', '', base)  # Remove anything in parentheses
    base = base.strip()
    return base

cars_df['base_model'] = cars_df['name'].apply(extract_base_model)

# Group by base model and count distinct years
model_years = cars_df.groupby('base_model')['model_year'].nunique()
multi_year_models = model_years[model_years > 1].index.tolist()

# Filter dataframe to only include models with multiple years
filtered_df = cars_df[cars_df['base_model'].isin(multi_year_models)]

# Task 2: For each model, compute std of mpg across years
model_stats = filtered_df.groupby('base_model').agg(
    appearances=('mpg', 'count'),
    avg_mpg=('mpg', 'mean'),
    std_mpg=('mpg', 'std'),
    years=('model_year', lambda x: sorted(x.unique()))
).reset_index()

# Task 3: Filter for models with std(mpg) < 1.0 and at least 2 appearances
consistent_models = model_stats[
    (model_stats['std_mpg'] < 1.0) & 
    (model_stats['appearances'] >= 2)
].sort_values(
    by=['appearances', 'avg_mpg'], 
    ascending=[False, False]
)

# Display the results
print("Most consistent car models (lowest MPG variation across years):")
print(consistent_models[['base_model', 'appearances', 'avg_mpg', 'std_mpg', 'years']].to_string(index=False))




```
